```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
headernames = ['Roll', 'Name', 'Beng', 'Eng', 'Math', 'Science', 'PEdu', 'WEdu', 'Comp', 'section']
```

```python
df=pd.read_csv("/content/drive/MyDrive/data_5_edited.csv" , names=headernames)

df.columns = headernames
df.drop(columns=['Roll', 'Name', 'Eng', 'Science', 'PEdu' , 'WEdu', 'Comp', 'section'], inplace=True)
df.shape
df.head()
```

```python
X = df.iloc[:, :].values
#y = df.iloc[:,1].values
```

```python
plt.scatter(X[:,0],X[:,1], color='black')
plt.show()
```

```python
from sklearn.cluster import KMeans

inertias = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, n_init='auto')
  kmeans.fit(X)
  inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2, n_init='auto')
kmeans.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c =kmeans.labels_)
plt.show()
```
